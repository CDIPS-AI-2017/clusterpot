```python
from collections import Counter
def people_info(doc):
    people = Counter()
    for ent in doc.ents:
        if ent.label_ in ('CARDINAL', 'ORDINAL'):
            continue
        elif ent.label_ == 'PERSON':
            people[ent.lemma_] += 1
    return people

    
```
